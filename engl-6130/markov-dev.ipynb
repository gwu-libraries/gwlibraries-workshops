```python
import re
import pandas as pd
from collections import defaultdict, Counter
from functools import reduce
```

```python
with open('./data/pg2600.txt') as f:
    text = f.read()
```

```python
data = re.split(r'\s+', text)
```

```python
transitions = defaultdict(lambda: defaultdict(list))
for word_1, word_2, next_word in zip(data[:-2], data[1:-1], data[2:]):
    transitions[word_1][word_2].append(next_word)
```

```python
def inc_counter(counter, word):
    counter[word] += 1
    return counter
for k1, v1 in transitions.items():
    for k2, words in v1.items():
        transitions[k1][k2] = reduce(inc_counter, words, Counter())
```

```python
transitions = {(k1, k2): v2 for k1, v1 in transitions.items() for k2, v2 in v1.items()}
```

```python
t_sorted = sorted([(k, v) for k,v in transitions.items()], key=lambda x: sorted(x[1].values(), key=None, reverse=True), reverse=True)
```

```python
corner = [(keys, dict(values.most_common(25))) for (keys, values) in t_sorted[:25]]
```

```python
corner ={keys: values for (keys, values) in corner}
```

```python
transitions_df = pd.DataFrame.from_dict(corner, orient='index')
```
